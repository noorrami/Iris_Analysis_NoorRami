# Task 1: Exploring and Visualizing the Iris

# Section 1: Environment Setup & Library Imports
We call upon the core libraries for data processing and visualization. We use Pandas for the software structure, while relying on Seaborn and Matplotlib to produce high-quality graphs that aid in visually understanding the data.

```python
# Import essential libraries for data science and visualization
import pandas as pd      # For data manipulation and analysis
import seaborn as sns   # For high-level statistical data visualization
import matplotlib.pyplot as plt # For basic plotting and customization

# Set the visualization style for aesthetic and professional-looking plots
sns.set_theme(style="whitegrid")
```

# Section 2: Dataset Acquisition & Initial Inspection
The popular Iris dataset is loaded. Here, we examine the data dimensions, column names, and data types (Dtypes). This step is essential to ensure there are no data type conflicts before beginning the analysis.

```python
# Load the Iris dataset directly from seaborn's built-in datasets
df = sns.load_dataset('iris')
```

```python
# Display basic information about the dataset structure and data types
print("--- Dataset Information ---")
df.info() # Checks for null values and data types for each column
print("Shape:", df.shape)
print("Column names:", list(df.columns))

# Display the first five rows to understand the feature values
print("\n--- First 5 Rows of the Dataset ---")
print(df.head())
```

# Section 3: Descriptive Statistical Analysis
We use the .describe() function to get a comprehensive view of the arithmetic mean, standard deviation, and minimum and maximum values. This helps us understand the scale of the data and determine if there is significant feature dispersion.

```python
# Display descriptive statistics to understand the distribution and scale
print("\n--- Statistical Summary ---")
print(df.describe()) # Provides mean, std, min, max, and quartiles
```

# Section 4: Relationship Visualization (Bivariate Analysis)
Using scatter plots, we analyze how the length and width of the petals and sepals are related. The goal is to determine which features are most effective at distinguishing the three types of flowers. It becomes clear that the petal features provide the strongest differentiation between the types.

```python
# Visualization 1: Scatter Plot to explore relationships between sepal length and width
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', ax=axes[0])
axes[0].set_title('Sepal Length vs Width')
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', ax=axes[1])
axes[1].set_title('Petal Length vs Width')
plt.tight_layout()
plt.show()
```

## Section 5: Feature Distribution & Quality Check
We plot histograms to understand the nature of the value distribution (such as a normal distribution). Additionally, we use box plots to detect outliers and examine the degree of overlap between different classes. We also ensure the data is free of missing values ​​to guarantee the accuracy of the analysis.

```python
# Visualization 2: Histograms to show the distribution of each numerical feature
# This helps in identifying the skewness and range of the data
df.hist(edgecolor='black', linewidth=1.2, figsize=(12, 10), color='skyblue')
plt.suptitle('Distribution of Numerical Features', fontsize=16)
plt.show()
```

```python
print("Missing values:\n", df.isnull().sum())
```

```python
# Visualization 3: Box Plot to identify outliers and compare distributions across species
plt.figure(figsize=(12,6))
sns.boxplot(data=df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
plt.title('Box Plots for All Features to Detect Outliers')
plt.show()
```
